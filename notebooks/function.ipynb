<h2>FONKSƒ∞YONLAR

def √∂n eki ile tanƒ±mlanƒ±r. 

args :  fonksiyona belirsiz sayƒ±da parametre alƒ±nmasƒ±nƒ± saƒülar (ba≈üƒ±na * koyularak atanƒ±r), t√ºm deƒüi≈ükenler aynƒ± tipte olmak zorunda, d√∂n√º≈ü deƒüeri tuple olur


kwargs : (ba≈üƒ±na ** koyularak atanƒ±r.), d√∂n√º≈ü deƒüeri dictionarydir, √ßoklu parametre alƒ±r.

python yazƒ±lƒ±m dilinde deƒüi≈ükenler ilk atanƒ±rken global olarak atanƒ±r (yani bir fonksiyon i√ßerisinde deƒüi≈ükeni atamamƒ±≈ü olmana raƒümen sayiyi kullanabilirsin)

```python
def convertToList(*params):
    list = []
    for i in params :
        list.append(i)
    return list
newList = convertToList("Ahmet", "Mehmet", "Ali")
newList[2]
```

```python
def cokluKey(**kwargs) :
    return kwargs
cokluKey(a = 25, b = 35)
    
```

```python
def masaSiparis(masaNumarasi, *siparisler):
    cesit=len(siparisler)
    print(f"{masaNumarasi} numarali masanin siparisleri : ")
    for i in siparisler:
        print(f" - {i}")
    print(f" {masaNumarasi} numarali masa {cesit} cesit siparis vermistir.")


masaSiparis(3, "ƒ∞stakoz", "Kebap", "D√∂ner")
masaSiparis(9, "Levrek", "Jumbo Karides", "≈ûalgam")
masaSiparis(7, "Meyve", "Tatli", "Kebapp")
```

```python
notlar = list(map(int, input("Ogrenci notlarini giriniz (virgulle ayirarak) : ").split(",")))
def notAnaliz(notlar):
    ortalama = sum(notlar)/len(notlar)
    print(f"Ogrenci notlarinin ortalamasi : {ortalama:.2f}")
    minNot = min(notlar)
    print(f"Ogrencinin en dusuk notu : {minNot}")
    maxNot = max(notlar)
    print(f"Ogrencinin en yuksek notu : {maxNot}")
notAnaliz(notlar)
```
